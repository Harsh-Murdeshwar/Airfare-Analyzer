```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
"""
blr to knu: data_collection_date: 10th sept, travel_date_start:10th sept
bom to knu: data_collection_date: 10th sept, travel_date_start:11th sept
knu to bom: data_collection_date: 10th sept, travel_date_start:11th sept 
"""
```

```python
# data_collection_date = datetime(2024, 10, 1)
# travel_date_start = datetime(2024, 10, 1)
data_collection_date = datetime(2024, 9, 10)
travel_date_start = datetime(2024, 9, 11)
Title = 'bom_to_knu'
```

```python
# Read the input CSV
file_name = Title + "_all_data.csv"
df = pd.read_csv(file_name, header=None)
```

```python
# Function to process each value in the row (now a column)
def process_value(value, travel_date):
    days_ago_str, price_str = value.split(' - â‚¹')
    if days_ago_str == 'Today':
        days_ago = 0
    else:
        days_ago = int(days_ago_str.split()[0])
        
    print(travel_date.weekday())
    
    booking_date = data_collection_date - timedelta(days=days_ago)
    price = price_str.replace(',', '')
    
    return [booking_date.strftime('%Y-%m-%d'), travel_date.strftime('%Y-%m-%d'), price]
```

```python
columns=['date_of_booking', 'date_of_travel', 'price']
data = []
```

```python
for i in range (len(df[0])):
    travel_date = travel_date_start + timedelta(days=i)
    for col in df.columns:
        if type(df[col][i]) is float:
            continue
        data_pt = process_value(df[col][i], travel_date)
        data.append(data_pt)
        
```

```python
df2 = pd.DataFrame(data, columns=columns)
```

```python
df2.to_csv('compiled_all_data_'+Title+'.csv', index=False)
```
