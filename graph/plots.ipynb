```python
import pandas as pd
```

```python
df = pd.read_csv("all.csv")
```

```python
data = list(df.values)
```

```python
final_data = {}
for row in data:
    key = row[0], row[1], row[2]
    val = int(row[3])
    if key in final_data:
        final_data[key] = max(final_data[key], val)
    else:
        final_data[key] = val
```

```python
final_data
```

```python
cleaned = []
for p in final_data:
    cleaned.append(list(p) + [final_data[p]])
 
headings = ["source", "destination", "day", "count"]
    
df = pd.DataFrame(cleaned,columns=headings)
```

```python
import numpy as np
```

```python
counts = np.zeros(7)
```

```python
DAYS = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
```

```python
for r in cleaned:
    ix = DAYS.index(r[2])
    counts[ix]+=r[3]
```

```python
counts
```

```python
import matplotlib
from matplotlib import pyplot as plt
```

```python
plt.bar(DAYS,height=counts)
ax = plt.gca()
ax.set_ylim([3200, 3350])
plt.title("Total Flights vs Day")
plt.xlabel("Day of Week")
plt.ylabel("Count")
plt.savefig("graph-flight-vs-day.png")
```
