```python
import pandas as pd
```

```python
df = pd.read_csv("all.csv")
```

```python
data = list(df.values)
```

```python
final_data = {}
for row in data:
    key = row[0], row[1], row[2]
    val = int(row[3])
    if key in final_data:
        final_data[key] = max(final_data[key], val)
    else:
        final_data[key] = val
```

```python
final_data
```

```python
cleaned = []
for p in final_data:
    cleaned.append(list(p) + [final_data[p]])
 
headings = ["source", "destination", "day", "count"]
    
df = pd.DataFrame(cleaned,columns=headings)
```
