```python
import csv
import datetime
import os
import re
import pandas as pd
```

```python
data_dir = "."
def get_files(file_re):
    files = []
    for filename in os.listdir(data_dir):
        print(filename)

def get_csv_data(filename):
    data = []
    with open(filename, 'r') as f:
        b = csv.reader(f)
        next(b)
        for row in b:
            data.append(row)
    return data
```
